```python
import pandas as pd
import re
```

```python
# 1. 데이터 로드 & 합치기
yuquiz = pd.read_csv("../data/yuquiz.csv")
runningman = pd.read_csv("../data/runningman.csv")
nolmyun = pd.read_csv("../data/nolmyun.csv")
```

```python
print(yuquiz.head())
```

```python
# program 컬럼 추가
yuquiz["program"] = "유퀴즈"
runningman["program"] = "런닝맨"
nolmyun["program"] = "놀면뭐하니"
```

```python
df = pd.concat([yuquiz, runningman, nolmyun])
```

```python
print(df.head())
print(df.shape)
```

```python
# 2. 텍스트 정제
def clean_text(text):
    text = str(text)

    # URL 제거
    text = re.sub(r"http\S+|www\S+", "", text)

    # 이모지 및 특수문자 제거
    text = re.sub(r"[^가-힣a-zA-Z0-9\s]", "", text)

    # 공백 정리
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
df["comment"] = df["comment"].apply(clean_text)

# 5글자 이하 댓글 제거
df = df[df["comment"].str.len() > 5]
```

```python
df[["program", "comment"]].head()
```

```python
# 3. 중복 댓글 제거
df.drop_duplicates(subset=["comment"], inplace=True)
```

```python
print(df.shape)
```

```python
# 4. 결측치 처리
df.dropna(subset=["comment"], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# 5. 저장
df.to_csv("../data/cleaned_comments.csv", index=False, encoding="utf-8-sig")
```
