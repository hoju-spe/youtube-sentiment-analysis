```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
#matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Windows
matplotlib.rcParams['font.family'] = 'AppleGothic'  # Mac
```

```python
df = pd.read_csv("/Users/user/youtube-sentiment-analysis/data/cleaned_comments_sentiment.csv")
```

```python
# program 컬럼으로 분리
programs = ["유퀴즈", "런닝맨", "놀면뭐하니"]
dfs = {p: df[df["program"] == p] for p in programs}
```

```python
# 감성 비율 시각화
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, (program, data) in enumerate(dfs.items()):
    counts = data["sentiment"].value_counts()
    colors = ["#FF6B6B", "#95E1D3", "#F8B500"]
    axes[i].pie(counts, labels=counts.index, autopct="%1.1f%%", colors=colors)
    axes[i].set_title(f"{program}\n댓글 감성 비율")

plt.tight_layout()
plt.savefig("/Users/user/youtube-sentiment-analysis/data/sentiment_comparison.png", dpi=150)
plt.show()
```
