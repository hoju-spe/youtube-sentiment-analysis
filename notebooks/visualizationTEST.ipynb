```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv("../data/yuquiz.csv")
# df = pd.read_csv("../data/runningman.csv")
# df = pd.read_csv("../data/nolmyun.csv")
df.head()
```

```python
text = " ".join(df["comment"].dropna().astype(str))

wc = WordCloud(
    font_path="C:/Windows/Fonts/malgun.ttf",
    width=1000,
    height=600,
    background_color="white"
).generate(text)

plt.figure(figsize=(12, 7))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
